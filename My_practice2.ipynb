# Завдання 3
Очистіть дані від викидів. Оберіть стовпчики на основі ящиків з вусами

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)  # Перший квартиль (25%)
    Q3 = df[column].quantile(0.75)  # Третій квартиль (75%)
    IQR = Q3 - Q1  # Міжквартильний розмах (IQR)

    lower_bound = Q1 - 1.5 * IQR  # Нижня межа
    upper_bound = Q3 + 1.5 * IQR  # Верхня межа


    df_cleaned = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_cleaned


columns_to_clean = ['Temperature (C)', 'Apparent Temperature (C)', 'Wind Speed (km/h)', 'Pressure (millibars)']


df_cleaned = df.copy()


for column in columns_to_clean:
    df_cleaned = remove_outliers(df_cleaned, column)


# print(df_cleaned.describe())
df_cleaned.shape
```
